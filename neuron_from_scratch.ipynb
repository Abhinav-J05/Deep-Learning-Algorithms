2 Layers of neurons. 

```python
# Layer of 4 Neurons

inputs = [1, 2, 3, 4]

weights = [
          [0.5, 0.3, 0.6, 0.7, 0.2],
          [0.1, 0.4, 0.2, 0.3, 0.4],
          [0.3, 0.5, 0.6, 0.2, 0.4],
          [0.4, 0.5, 0.2, 0.8, 0.3],
          ]

biases = [1, 1, 2, 1]

output1 = inputs[0]*weights[0][0] + inputs[1]*weights[0][1] + inputs[2]*weights[0][2] + inputs[3]*weights[0][3]
output2 = inputs[0]*weights[1][0] + inputs[1]*weights[1][1] + inputs[2]*weights[1][2] + inputs[3]*weights[1][3]
output3 = inputs[0]*weights[2][0] + inputs[1]*weights[2][1] + inputs[2]*weights[2][2] + inputs[3]*weights[2][3]
output4 = inputs[0]*weights[3][0] + inputs[1]*weights[3][1] + inputs[2]*weights[3][2] + inputs[3]*weights[3][3]

output  = [output1, output2, output3, output4]

print(output)
```

3. Layer of nerons using loops

```python
inputs = [1, 2, 3, 4]

weights = [
          [0.5, 0.3, 0.6, 0.7, 0.2],
          [0.1, 0.4, 0.2, 0.3, 0.4],
          [0.3, 0.5, 0.6, 0.2, 0.4],
          [0.4, 0.5, 0.2, 0.8, 0.3],
          ]

biases = [1, 1, 2, 1]

output = []

for  weight, bias in zip(weights, biases):

    result = 0
    for w, inp in zip(weight, inputs):
        result = w*inp + result

    output.append(result + bias)

print(output)
```
